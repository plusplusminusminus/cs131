```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('generated_data.csv')
```

```python
pivoted_df = pd.pivot_table(df, index='Year', columns='Country/Territory', values='Total Organ-Related Casualties')
pivoted_df
```

```python
plt.figure(figsize=(12,8))
plt.title('Total Organ-Related Casualties (Top 10 countries with the highest casualties)')
plt.xlabel('Year')
plt.ylabel('Casualties')
plt.ticklabel_format(style='plain')
plt.plot(pivoted_df.index, pivoted_df, label=pivoted_df.columns)
plt.legend()
```
